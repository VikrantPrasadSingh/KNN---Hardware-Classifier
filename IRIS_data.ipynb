```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
```

```python
df.head()
```

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler

iris = load_iris()
X = iris.data
y = iris.target

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)


print(X_scaled[:5])
```

```python
df.head()
```

```python
X = df.drop('target', axis = 1)
Y = df['target']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size = 0.2, random_state = 42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model_KNN = KNeighborsClassifier()
model_KNN.fit(X_train_scaled, y_train)
```

```python
y_pred = model_KNN.predict(X_test_scaled)
```

```python
y_pred
```

```python
accuracy_score(y_pred, y_test)
```

```python
# Export training data (scaled)
np.savetxt("X_train_scaled.csv", X_train_scaled, delimiter=",")
np.savetxt("y_train.csv", y_train, delimiter=",", fmt="%d")

# Export test data (scaled)
np.savetxt("X_test_scaled.csv", X_test_scaled, delimiter=",")
np.savetxt("y_test.csv", y_test, delimiter=",", fmt="%d")

# Export predictions
np.savetxt("y_pred.csv", y_pred, delimiter=",", fmt="%d")

print("All files exported!")
```

```python
import numpy as np

# X fixed point conversion
X_train_fixed = (X_train_scaled * 255).round().astype(int)
X_test_fixed = (X_test_scaled * 255).round().astype(int)

# Export X
np.savetxt("X_train_fixed.csv", X_train_fixed, delimiter=",", fmt="%d")
np.savetxt("X_test_fixed.csv", X_test_fixed, delimiter=",", fmt="%d")

# Export Y (labels) - already integers, no fixed point needed
np.savetxt("y_train.csv", y_train, delimiter=",", fmt="%d")
np.savetxt("y_test.csv", y_test, delimiter=",", fmt="%d")

# Export predictions (golden reference for Verilog verification)
np.savetxt("y_pred.csv", y_pred, delimiter=",", fmt="%d")

print("X_train_fixed shape:", X_train_fixed.shape)
print("X_test_fixed shape:", X_test_fixed.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```
